## Task 4: Experiment With The Temperature Parameter

```python
def get_response(prompt, generation_config={}):
    response = model.generate_content(contents=prompt, 
    generation_config=generation_config)
    return response

for temp in [0.0, 0.25, 0.5, 0.75, 1.0]:
  config = genai.types.GenerationConfig(temperature=temp)
  result = get_response("Explain the concepts of XGBoost and Random Forest with real-life use cases", generation_config=config)

  print(f"\n\nFor temperature value {temp}, the results are: \n\n")
  display(Markdown(result.text))
```

## Task 10: Load the PDF and Extract the Texts

```python
CHUNK_SIZE = 700
CHUNK_OVERLAP = 100
pdf_path = "http://jmc.stanford.edu/articles/whatisai/whatisai.pdf"
```

```python
if PyPDFLoader:
    try:
        pdf_loader = PyPDFLoader(pdf_path)
        split_pdf_document = pdf_loader.load_and_split()
        print(f"✅ Successfully loaded PDF with {len(split_pdf_document)} pages")
    except Exception as e:
        print(f"❌ Error loading PDF: {e}")
        print("💡 Alternative: You can manually copy-paste text or use a different approach")
        # Create dummy documents for demonstration
        split_pdf_document = []
else:
    print("❌ PyPDFLoader not available")
    print("💡 Installing required package...")
    print("Run: pip install langchain-community")
    
    # For demonstration, create some sample text
    sample_text = """
    Artificial Intelligence (AI) refers to the simulation of human intelligence in machines 
    that are programmed to think like humans and mimic their actions. The term may also be 
    applied to any machine that exhibits traits associated with a human mind such as learning 
    and problem-solving.
    
    Machine learning is a subset of artificial intelligence (AI) that provides systems the 
    ability to automatically learn and improve from experience without being explicitly 
    programmed. Machine learning focuses on the development of computer programs that can 
    access data and use it to learn for themselves.
    """
    
    # Create dummy document objects
    class SimpleDocument:
        def __init__(self, content):
            self.page_content = content
    
    split_pdf_document = [SimpleDocument(sample_text)]
    print("📄 Using sample text for demonstration")
```

```python
if RecursiveCharacterTextSplitter and split_pdf_document:
    # Splitting text into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=CHUNK_SIZE, chunk_overlap=CHUNK_OVERLAP)
    context = "\n\n".join(str(p.page_content) for p in split_pdf_document)
    texts = text_splitter.split_text(context)
    print(f"✅ Successfully split text into {len(texts)} chunks")
else:
    print("❌ Text splitter not available or no documents loaded")
    
    # Simple fallback text splitting
    if split_pdf_document:
        context = "\n\n".join(str(p.page_content) for p in split_pdf_document)
        # Simple chunking by sentences
        sentences = context.split('.')
        texts = []
        current_chunk = ""
        
        for sentence in sentences:
            if len(current_chunk + sentence) < CHUNK_SIZE:
                current_chunk += sentence + "."
            else:
                if current_chunk:
                    texts.append(current_chunk.strip())
                current_chunk = sentence + "."
        
        if current_chunk:
            texts.append(current_chunk.strip())
        
        print(f"📝 Using simple text splitting: {len(texts)} chunks")
    else:
        texts = ["Sample text for demonstration"]
        print("📝 Using sample text chunks")

print(f"\n📊 Text Chunks Summary:")
print(f"Total chunks: {len(texts)}")
print(f"Average chunk length: {sum(len(t) for t in texts) // len(texts) if texts else 0} characters")
print(f"First chunk preview: {texts[0][:100]}..." if texts else "No chunks available")
```
